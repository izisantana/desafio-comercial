**Tratando as datas**

Para análise temporal, separo as datas em colunas.

```python
vendas_total["Date"] = pd.to_datetime(vendas_total["Date"])
vendas_total["Year"] = vendas_total["Date"].dt.year
vendas_total["Month"] = vendas_total["Date"].dt.month
vendas_total["Quarter"] = "T" + vendas_total["Date"].dt.quarter.astype(str)
vendas_total["Week"] = vendas_total["Date"].dt.isocalendar().week.astype(int)
```

Criei a coluna de quarter, porque eu já fiz essa tratativa de dados de data algumas vezes e quando eu montei o dash, pensei em olhar por trimestre no filtro, em vez de semana, mas acabei olhando o mês, deixei ela na planilha porque pooderia ser útil.
